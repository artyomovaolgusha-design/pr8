```python
def inn():
    while True:
        x = input("ИНН (12 цифр): ")
        if len(x) == 12 and x.isdigit():
            print("инн принят")
            return x
        print("Ошибка, ввидете инн еще раз")

inn()
```

```python
def price():
    while True:
        try:
            x = float(input("Цена: "))
            if x > 0 and x <= 1000000:
                print("")
                return x
            print("Диапазон 0-1 млн")
        except:
            print("Введите число!")

price()
```

```python
def password():
    while True:
        p = input("Пароль: ")
        if len(p) < 8:
            print("Короткий пароль, ввидите заново")
            continue
        if len(p) > 20:
            print("Длинный, ввидите заново")
            continue

        big = False
        num = False
        for c in p:
            if c.isupper(): big = True
            if c.isdigit(): num = True

        if big and num:
            print("Готово, ваш пароль:")
            return p
        print("Нужна заглавная буква и цифра")

password()
```
